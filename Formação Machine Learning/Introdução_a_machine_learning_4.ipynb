```python
import pandas as pd

uri = 'https://gist.githubusercontent.com/guilhermesilveira/4d1d4a16ccbf6ea4e0a64a38a24ec884/raw/afd05cb0c796d18f3f5a6537053ded308ba94bf7/car-prices.csv'
dados = pd.read_csv(uri)
dados.head()
```

```python
a_renomear = {
    "mileage_per_year" : "milhas_por_ano",
    "model_year" : "ano_do_modelo",
    "price" : "preco",
    "sold" : "vendido"
}
dados = dados.rename(columns=a_renomear)
dados.head()
```

```python
troca = {
    "no" : 0,
    "yes" : 1
}
dados.vendido = dados.vendido.map(troca)
dados.head()
```

```python
from datetime import datetime

ano_atual = datetime.today().year

dados["idade_do_modelo"] = ano_atual - dados.ano_do_modelo
dados.head()
```

```python
dados["km_por_ano"] = dados.milhas_por_ano * 1.60934
dados.head()
```

```python
dados = dados.drop(columns=["Unnamed: 0", "milhas_por_ano", "ano_do_modelo"], axis=1)
dados.head()
```

```python
x = dados[["preco", "idade_do_modelo", "km_por_ano"]]
y = dados["vendido"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
import numpy as np

# semente usada para separar dados de treino e dados de teste
SEED = 5
# define a seed usada pelo numpy pra gerar o número aleatório
np.random.seed(SEED)

# por padrão, separa treino e teste aleatoriamente...
treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, 
                                                        test_size = 0.25,
                                                        stratify = y)
print("treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x), len(teste_x)))

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("Taxa de acerto: %.2f%%" % acuracia)
```
